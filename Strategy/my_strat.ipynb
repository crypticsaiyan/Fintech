**Basic Trading Strategy using Python**

```python
from curl_cffi import requests
session = requests.Session(impersonate="chrome")

import yfinance as yf
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
# rsi, macd, sma, ema, bb
def rsi(data, period=20):
    data['PD'] = data['Close'].diff()
    data['P'] = data['PD'].apply(lambda x: x if x>0 else 0)
    data['L'] = data['PD'].apply(lambda x: abs(x) if x<0 else 0)
    data['avgp'] = data['P'].rolling(window=period, min_periods=1).mean()
    data['avgl'] = data['L'].rolling(window=period, min_periods=1).mean()
    data['RS'] = data['avgp']/(data['avgl'] + 1e-10)
    data['RSI'] = 100 - (100/(1 + data['RS']))
    return data

def macd(data, short_period=12, long_period=26, signal_period=9):
    data['ema12'] = data['Close'].ewm(span=short_period, adjust=False).mean()
    data['ema26'] = data['Close'].ewm(span=long_period, adjust=False).mean()
    data['MACD'] = data['ema12'] - data['ema26']
    data['SIG'] = data['MACD'].ewm(span=signal_period, adjust=False).mean()
    return data

def sma(data, period=20):
    data['SMA'] = data['Close'].rolling(window=period, min_periods=1).mean()
    return data

def ema(data, period=20):
    data['EMA'] = data['Close'].ewm(span=period, adjust=False).mean()
    return data

def bb(data, period=20):
    data['MB'] = data['Close'].rolling(window=period, min_periods=1).mean() #middle band
    data['STD'] = data['Close'].rolling(window=period, min_periods=1).std()
    data['UB'] = data['MB'] + 2 * data['STD']; #upper band
    data['LB'] = data['MB'] - 2 * data['STD']; #lower band
    return data.dropna()

#some extra indicators
def atr(data, period=14): #avg true range
    data = data.copy()
    data['O1'] = data['High'] - data['Low'] 
    data['O2'] = (data['High'] - data['Close'].shift(1)).abs() 
    data['O3'] = (data['Low'] - data['Close'].shift(1)).abs()
    data['TR'] = data[['O1', 'O2', 'O3']].max(axis=1)
    
    # data['ATR'] = data['TR'].ewm(span=period, adjust=False).mean()
    data['ATR'] = data['TR'].rolling(window=period, min_periods=1).mean()

    for i in range(period, len(data)):
        prev_atr = data['ATR'].iloc[i-1]
        current_tr = data['TR'].iloc[i]
        data.at[data.index[i], "ATR"] = (prev_atr * (period - 1) + current_tr)/period

    return data

def adx(data, period=14): #avg directional index
    # calculate the directional movement
    data['DM+'] = data['High'] - data['High'].shift(1)
    data['DM-'] = data['Low'].shift(1) - data['Low']
    # condition, iftrue, iffalse
    data['DM+'] = np.where((data['DM+']> 0) & (data['DM+']>data['DM-']), data['DM+'], 0)
    data['DM-'] = np.where((data['DM-']>0 )& (data['DM-']> data['DM+']), data['DM-'], 0)

    data['sDM+'] = data['DM+'].ewm(span=period, adjust=False).mean()
    data['sDM-'] = data['DM-'].ewm(span=period, adjust=False).mean()
    data['sTR'] = data['TR'].ewm(span=period, adjust=False).mean()
    #directional indicators
    data['+DI'] = (data['sDM+']/data['sTR'])*100
    data['-DI'] = (data['sDM-']/data['sTR'])*100

    #directional index
    data['DX'] = (((data['+DI'] - data['-DI']).abs())/(data['+DI']+data['-DI']))*100
    data['ADX'] = data['DX'].ewm(span=period, adjust=False).mean()
    return data

def strat(data):
    data['Strat'] = 0
    flag = 0
    # long open = 1
    # long close = -1
    # short open = 2
    # short close = -2
    # do nothing = 0
    # i have used rsi, macd and bb in this strategy
    for i in range(len(data)):
        # go long if close < lower band or rsi < 30 or macd line > signal line and adx>25 and +di> -di
        if flag==0 and (((data['Close'].squeeze().iloc[i] <=  data['LB'].iloc[i]) | (data['RSI'].iloc[i] < 30) | (data['MACD'].iloc[i] > data['SIG'].iloc[i]))&((data['ADX'].iloc[i]>25)& (data['+DI'].iloc[i]> data['-DI'].iloc[i]))):
            data.at[data.index[i], "Strat"] = 1
            flag=1
        # exit when rsi > 65 or close > middle band or +di < -di
        # exit if loss becomes 15% of the long posi
        elif flag==1 and (((data['RSI'].iloc[i]>65) & (data['Close'].squeeze().iloc[i] >= data['MB'].iloc[i]) ) | ((data['ADX'].iloc[i]>25)&(data['+DI'].iloc[i]<data['-DI'].iloc[i]))):
            data.at[data.index[i], "Strat"] = -1
            flag = 0
        # go short if close > upper band or rsi > 70 or macd line < signal line and adx>25 and +di < -di
        elif flag==0 and (((data['Close'].squeeze().iloc[i] >=  data['UB'].iloc[i]) | (data['RSI'].iloc[i] > 70) | (data['MACD'].iloc[i] < data['SIG'].iloc[i]))&((data['ADX'].iloc[i]>25)& (data['+DI'].iloc[i]< data['-DI'].iloc[i]))):
            data.at[data.index[i], "Strat"] = 2
            flag=2
        # exit when rsi < 35 or close < middle band or +di > -di
        elif flag==2 and (((data['RSI'].iloc[i]<35) & (data['Close'].squeeze().iloc[i] <= data['MB'].iloc[i]) ) | ((data['ADX'].iloc[i]>25)&(data['+DI'].iloc[i]>data['-DI'].iloc[i]))):
            data.at[data.index[i], "Strat"] = -2
            flag = 0                    
    return data

# calculate the profit after following the above strategy
def calc_profit(data):
    # if 1 then find -1 -> if found -> prof+= close(index at which -1 was found) - close(ind at 1)
    #                      else -> prof+= close(last) - close(ind at 1)
    # if 2 then find -2 -> if found -> prof+= close(index at which 2 was found) - close(ind at -2)
    #                      else -> prof+= close(2) - close(ind at last)
    data = data.copy()
    profit = 0
    prev = 0
    for i in range(len(data)):
        if data['Strat'].iloc[i] == 1:
            prev = data['Close'].iloc[i]
        elif data['Strat'].iloc[i] == -1:
            profit = profit + data['Close'].iloc[i] - prev
        elif data['Strat'].iloc[i] == 2:
            prev = data['Close'].iloc[i]
        elif data['Strat'].iloc[i] == -2:
            profit = profit - data['Close'].iloc[i] + prev
    return profit

tickers = ["NVDA", "AAPL", "NFLX", "GOOGL", "TSLA"]
# tickers = ["TSLA"]
start_day = "2025-05-01"
end_day = datetime.now().strftime("%Y-%m-%d")

all_data = []

total_prof = 0
for ticker in tickers:
    try:
        data = yf.download(ticker, start=start_day, end=end_day, session=session)
        data = rsi(data)
        data = macd(data)
        data = sma(data)
        data = ema(data)
        data = bb(data)
        data = atr(data)
        data = adx(data)
        data = strat(data)

        data["Ticker"] = ticker
        all_data.append(data)

        print(data[['Close', 'RSI', 'MACD','SIG', 'SMA', 'EMA','MB', 'LB', 'UB','ATR', 'ADX', 'Strat']].round(2))
        # print(data[['Close', 'RSI', 'MACD','SIG','MB', 'LB', 'UB', 'ADX', 'Strat']].tail().round(2))

        # Using AI to visualize data
        # Create visualization
        plt.figure(figsize=(8, 4))
        plt.plot(data['Close'], label='Close Price', color='blue')
        plt.title(f'{ticker} Closing Price with Indicators')
        plt.xlabel('Date')
        plt.ylabel('Price ($)')

        # Add Bollinger Bands to the plot
        plt.plot(data['UB'], label='Upper Band', linestyle='--', color='red')
        plt.plot(data['MB'], label='Middle Band', linestyle='--', color='orange')
        plt.plot(data['LB'], label='Lower Band', linestyle='--', color='green')
        
        # Add markers for strategy signals
        plt.scatter(data[data['Strat'] == 1].index, 
                   data[data['Strat'] == 1]['Close'], 
                   marker='^', color='lime', label='Long Entry', zorder=5)
        plt.scatter(data[data['Strat'] == -1].index, 
                   data[data['Strat'] == -1]['Close'], 
                   marker='v', color='darkgreen', label='Long Exit', zorder=5)
        plt.scatter(data[data['Strat'] == 2].index, 
                   data[data['Strat'] == 2]['Close'], 
                   marker='v', color='red', label='Short Entry', zorder=5)
        plt.scatter(data[data['Strat'] == -2].index, 
                   data[data['Strat'] == -2]['Close'], 
                   marker='^', color='darkred', label='Short Exit', zorder=5)
        
        plt.legend()
        plt.grid(True)
        plt.show()

        # print(data[['Close', 'High', 'Low', 'DM+', 'DM-','TR', '+DI', '-DI','DX', 'ADX', 'Strat']].tail().round(2))
        # print(data[['Close', 'RSI', 'MACD', 'SMA', 'EMA', 'LB', 'UB']].head(16).round(2))
        # print(type(data['Close'].squeeze()))
        # print(type(data['RSI']))

        # print(data['Close'].squeeze()[281])
        # print(data['Close'].squeeze()))
        # print("\n")
        print(f"Profit:\n {calc_profit(data)}")
        total_prof = total_prof + float(calc_profit(data).iloc[0])
        print("\n-------------------------------------------------------------------------------------\n")
        # break
    except:
        print("Download failed!")
        # break
    time.sleep(1)

print("Total Profit: ")
print(round(total_prof, 2))

if all_data:
    combined_df = pd.concat(all_data)
    combined_df.to_csv("all_stocks_strategy_results.csv")
    print("Saved all stocks strategy results to all_stocks_strategy_results.csv")


# drawdown, sharpe
```
